# Problema

Como hotel con alta demanda, queremos anticipar si una reserva será cancelada, para minimizar pérdidas económicas y optimizar la asignación de habitaciones.

Trabajaremos con un conjunto de datos de reservas de hoteles y nuestro objetivo sera predecir si la reserva será cancelada o no.

Dado que nuestra variable objetivo es un valor discreto binario (0=no cancela | 1=cancela), entonces estamos frente a un problema de clasificación.


# Metrica

Para evaluar y comparar los distintos modelos utilizaremos las métricas $F1 Score$ y $Recall$. Tomamos esta decisión ya que priorizamos el rendimiento economico del hotel (a traves del Recall) pero sin descuidar su reputación (al buscar un buen valor de F1 Score, que tiene en cuenta la $Precision$).



# Pasos

1. Auto-EDA: Exploración automatizada preliminar. Se realiza un análisis rápido y automático del dataset para obtener una primera impresión general.

2. EDA: Exploración de datos. Se analiza el conjunto de datos en profundidad para entender sus propiedades principales. Esta etapa utilizaremos herramientas visuales (gráficos, distribuciones, correlaciones), aunque también puede incluir métodos no visuales.

3. Data-Cleaning (Limpieza de datos): En esta fase se corrigen o eliminan registros problemáticos detectados durante el EDA, como datos repetidos, ausentes o inconsistentes.

4. Modelado: Se eligen y configuran los algoritmos de aprendizaje, se entrena el modelo con los datos y luego se evalúa su desempeño utilizando métricas apropiadas (mencionadas anteriormente).
